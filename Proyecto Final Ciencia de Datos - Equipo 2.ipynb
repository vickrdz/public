#### Criterios de recategorización de la variable objetivo *loan_status*

**Definición del criterio de apetito de riesgo a utilizar**:

- Habiendo analizado los datos, y los posibles valores de la variable predictora "loan_status", decidimos mantener un perfil de riesgo bajo en el que tengamos la certeza de que podamos aceptar solo a los clientes que tengan una alta probabilidad de pagarnos.
Por lo tanto, excluímos los valores que no reflejan determinantemente la conducta de pago, es decir, aquéllos diferentes a 'Fully paid' (sí pagan), y diferentes a 'Charged off' (no pagan). Por lo tanto, los registros utilizados son solo los que cumplan con esos dos valores, asumiento que los demás aún tienen posibilidad de pagar o de no hacerlo, sin la certeza de que ésto haya ocurrido. 
Además, en el análisis exploratorio, pudo observarse la gran cantidad de observaciones para el valor 'Current', lo cual pondría en riesgo la fiabilidad del modelo, en caso de clasificar este valor como "positivo" o "negativo".


**Etiquetar los datos de la variable objetivo en 3 grupos de manera que permita validar el nivel de riesgo**

- **Grupo 0 -  Nivel de riesgo alto: 0**
- **Grupo 1 -  Nivel de riesgo bajo: 3**
- **Grupo 3 -  Grupo a eliminarse: 1, 2, 4, 5 y 6**

#### Balancear las etiquetas

```python
from sklearn.utils import resample
#Crear dos diferentes dataframe de una clase mayoritaria y minoritaria
df_majority = df2[(df2['loan_status']==1)]
df_minority = df2[(df2['loan_status']==0)]

# muestreo ascendente de la clase minoritaria
df_minority_upsampled = resample(df_minority,
                                 replace=True,     # muesta con reemplazo
                                 n_samples= 417499, # para que coincida con la clase mayoritaria
                                 random_state=0)   # resultados reproducible

# Combinar la clase mayoritaria con la muestra ascendente de la clase minoritaria
df_upsampled = pd.concat([df_minority_upsampled, df_majority])
```

```python
sns.countplot(x= 'loan_status', data = df_upsampled)
df_upsampled['loan_status'].value_counts()
```

#### Eliminar valores atípicos IQR

```python
def remove_outliers_iqr(df3, columns):
    for col in columns:
        q1 = df3[col].quantile(0.25)
        q3 = df3[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df2 = df3[(df3[col] >= lower_bound) & (df3[col] <= upper_bound)]
    return df3

# Señale las columnas para remover los outliers
columns_to_check = ['fico_range_high', 'fico_range_low','inq_last_6mths', 'int_rate', 'open_acc_6m']

# Solicitar la función que remueve los outliers usando IQR
df_clean = remove_outliers_iqr(df_upsampled, columns_to_check)

# Mostrar el resultado en el dataframe
df_clean.head()
```

## Algoritmos de Predicción

### Justificación del Porcentaje de División 70-30 para la Creación de los Datasets

La división de los datos en un 70% para entrenamiento y un 30% para testeo se basa en varios criterios fundamentales:

1. **Balance entre Entrenamiento y Validación:**
   - Utilizar el 70% de los datos para entrenamiento asegura que el modelo tenga suficiente información para aprender patrones significativos.
   - El 30% restante para testeo proporciona un conjunto de datos lo suficientemente grande como para evaluar el rendimiento del modelo de manera robusta y evitar el sobreajuste.

2. **Generalización del Modelo:**
   - Esta proporción permite evaluar cómo se comporta el modelo con datos no vistos, asegurando que las predicciones sean generalizables a nuevos datos.

3. **Prácticas Comunes:**
   - La división 70-30 es una práctica común en machine learning que equilibra adecuadamente la necesidad de datos para entrenamiento y validación.

4. **Eficiencia Computacional:**
   - Este porcentaje es eficiente en términos de tiempo y recursos computacionales, permitiendo entrenar y validar modelos sin requerir cantidades excesivas de datos o tiempo de procesamiento.

### Utilizar el método de K-FOLD CROSS VALIDATION

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Create a logistic regression model
model = LogisticRegression()

# Create a KFold object with desired number of folds
# n_splits: number of cross-validation splits
# shuffle: boolean indicating to shuffle data before splitting
# random_state: random seed
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform k-fold cross-validation
scores = cross_val_score(model, X, y, cv=kfold)

# Print the mean accuracy across all folds
print("Mean accuracy:", scores.mean())

#To create just the index 
splits = kfold.split(X)
```
